```python
import pandas as pd
```

# General lecture data preparation with Python

```python
df = pd.read_csv("MockPharmacyData.csv")
df.head()
```

```python
df['PatientID'] = df['PatientID'].str.replace('-','').astype('int')
df.head()
```

```python
df["DispensedDate"] = pd.to_datetime(df["DispensedDate"]).dt.strftime('%Y-%m-%d')
df.drop(columns='DispensedDateTime', inplace=True)
df.head()
```

```python
df2 = df.iloc[:,[0,1,2,3,4]]
df2.to_csv("MockPatientData.csv", index=False)
```

```python
df = df.drop(labels=['FirstName', 'LastName', 'Email', 'Gender'], axis=1)
df.head()
```

```python
df.to_csv("MockPharmacyDispenseData.csv", index=False)
```
